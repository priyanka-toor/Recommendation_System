```python
import pandas as pd
s_data=pd.read_excel("subset_data.xlsx")
s_data.info()
```

```python
#2
invoice_data = s_data.groupby('InvoiceNo')['StockCode'].apply(list).reset_index()
invoice_data.columns = ['InvoiceNo', 'StockCodes']
print(invoice_data)
```

```python
#3
# Group data by InvoiceNo and aggregate StockCodes as a list
#the lambda function inside the apply() method is used to ensure that all 'StockCode' values are treated as strings.
invoice_data = s_data.groupby('InvoiceNo')['StockCode'].apply(lambda x: list(map(str, x))).reset_index()
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
# Convert StockCodes into one-hot encoded format
te = TransactionEncoder()
te_ary = te.fit(invoice_data['StockCode']).transform(invoice_data['StockCode'])
df = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
frequent_itemsets = apriori(df, min_support=0.01, use_colnames=True)

# Generate association rules with lift more than 1.0
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.0)
```

```python
print(rules)
```

```python
#3
# Group data by InvoiceNo and aggregate StockCodes as a list
#the lambda function inside the apply() method is used to ensure that all 'StockCode' values are treated as strings.
invoice_data2= s_data2.groupby('InvoiceNo')['StockCode'].apply(lambda x: list(map(str, x))).reset_index()

from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
# Convert StockCodes into one-hot encoded format
te = TransactionEncoder()
te_ary = te.fit(invoice_data2['StockCode']).transform(invoice_data2['StockCode'])
df2 = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
# Generate frequent itemsets with minimum support of 0.01
from scipy.sparse import csr_matrix

# Convert the DataFrame to a sparse matrix
#sparse_df = csr_matrix(df2.values)

# Generate frequent itemsets with sparse matrix
frequent_itemsets = apriori(df2,min_support=0.01, use_colnames=True)

# Generate association rules with lift more than 1.0
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.0)
```
